## Types

- Inner Join → Only the overlap (3, 4)
- Left Join → Entire left circle (1, 2, 3, 4)
- Right Join → Entire right circle (3, 4, 5, 6)
- Outer Join → Both circles together (1, 2, 3, 4, 5, 6)

```python
import pandas as pd

data = pd.read_excel("test_data.xlsx")

data
```

```python
data1 = pd.read_excel("modify.xlsx")

data1
```

```python
pd.merge(data , data1)
```

```python
# inner join 

pd.merge(data , data1, how = 'inner')
```

```python
pd.merge(data , data1, how = 'outer')
```

```python
df3 = pd.read_excel("df3.xlsx")
df3
```

```python
pd.merge(df3 , data1 , left_on = "name" , right_on = "sales_man_name").drop("sales_man_name", axis = 1)
```

```python
pd.merge(data , data1, how = "left")
```

```python
pd.merge(data1 , data, how = "left")
```

```python
pd.merge(data , data1, how = "right")
```
