```python
data.rename(columns={"sex":"Gender"} , inplace=True)  
```

```python
data.age.fillna(data.age.median(), inplace = True)
```

```python
data["class"].value_counts()
```

```python
sns.countplot(x=data["class"], palette=["r" , "g", "y"])
```

```python
sns.countplot(x=data["class"], palette="rainbow")
```

```python
plt.title("Class Counts")
sns.countplot(x=data["class"], palette="rainbow")
plt.xlabel("Classes Name")
plt.ylabel("Count of Classes")
plt.show()
```

# Pieplot

```python
x = data.embark_town.value_counts().values
y = data.embark_town.value_counts().index
```

```python
plt.pie(x=x)
```

```python
plt.title("Pie Chart of Embarked Town", fontdict={"size":15, "color":"Orange"})
plt.pie(x=x , autopct="%.2f%%", labels=y, colors = ["y", "b", "g"])
plt.show()
```
