## Computing indicators / Dummy Variables

# numerical =  2 , 1 , 3 , 4

# categorical data = manager , male , female , yes , no

## categorical ---- ordinal ---- nomimal

### Nominal data 

Examples:

Blood Group: A, B, AB, O
Gender: Male, Female
Eye Color: Brown, Blue, Black
City: Karachi, Lahore, Islamabad

You cannot say:

Karachi > Lahore ❌
Blue > Brown ❌

There is no meaningful order.


### Ordinal data


Examples:

Education: High School → Bachelor's → Master's → PhD
Satisfaction: Very Dissatisfied → Dissatisfied → Neutral → Satisfied → Very Satisfied
T-Shirt Size: Small → Medium → Large → XL
Movie Rating: 1 Star → 2 Stars → 3 Stars → 4 Stars → 5 Stars
Grades: A+ → A → B+ → B → C → D

You know:

Large > Medium > Small ✅

But you don't know how much larger one category is than another.
