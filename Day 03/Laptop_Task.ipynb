```python
import pandas as pd 

df = pd.read_csv("laptop.csv")
```

```python
df
```

```python
df.fillna({"Unnamed: 0": df["Unnamed: 0"].mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df.fillna({"Company": df.Company.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df.fillna({"TypeName": df.TypeName.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df["Inches"] = df["Inches"].str.replace('"', "", regex=False)
```

```python

df["Inches"] = df["Inches"].replace({"none": None, "?": None})
```

```python
df["Inches"] = pd.to_numeric(df["Inches"], errors="coerce")
```

```python
df.fillna({"Inches": df.Inches.mode()[0]}, inplace=True)

df.isna().sum()

df.info()
```

```python
df.fillna({"ScreenResolution": df.ScreenResolution.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df.fillna({"Cpu": df.Cpu.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df.info()
```

```python
df["Ram"] = df["Ram"].str.replace('"', "", regex=False)
df["Ram"] = df["Ram"].str.replace("GB", "", regex=False)
```

```python
df["Ram"] = pd.to_numeric(df["Ram"], errors="coerce")
df.fillna({"Ram": df.Ram.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df["Memory"] = df["Memory"].replace({"?": None})
df.fillna({"Memory": df.Memory.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df.fillna({"Gpu": df.Gpu.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df
```

```python
df["OpSys"] = df["OpSys"].replace({"Mac OS X": "macOS", "Windows 10 S": "Windows 10"})
df.fillna({"OpSys": df.OpSys.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df["Weight"] = df["Weight"].str.replace("kg", "", regex=False)
df["Weight"] = df["Weight"].replace({"?": None})
```

```python
df["Weight"] = pd.to_numeric(df["Weight"], errors="coerce")
df.fillna({"Weight": df.Weight.mode()[0]}, inplace=True)

df.isna().sum()
```

```python
df["Price"] = df["Price"].replace({"none": None})
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df.fillna({"Price": df.Price.mode()[0]}, inplace=True)

df.isna().sum()
```
